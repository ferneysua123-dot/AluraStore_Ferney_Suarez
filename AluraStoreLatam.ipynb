### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
Facturacion1 = tienda1['Precio'].sum()
Facturacion2 = tienda2['Precio'].sum()
Facturacion3 = tienda3['Precio'].sum()
Facturacion4 = tienda4['Precio'].sum()


```

```python
print(f"El ingreso total en tienda 1 = ${Facturacion1:,.0f}")
print(f"El ingreso total en tienda 2 = ${Facturacion2:,.0f}")
print(f"El ingreso total en tienda 3 = ${Facturacion3:,.0f}")
print(f"El ingreso total en tienda 4 = ${Facturacion4:,.0f}")
```

```python
# lista con DataFrames
sucursales = [tienda1, tienda2, tienda3, tienda4]

# Diccionario para almacenar las ventas
facturacion = {}

# Iterar sobre la lista de DataFrames
for i, tienda in enumerate(sucursales):
    # Calculamos el promedio de la columna 'Calificación'
    ventas = tienda['Precio'].sum()
    # Almacenamos el promedio en el diccionario, usando un nombre descriptivo
    facturacion[f'tienda {i+1}'] = ventas

# Imprimimos los promedios
for nombre_tienda, ventas in facturacion.items():
    print(f'las ventas en {nombre_tienda}: {ventas:,.0f}')
```

# 2. Ventas por categoría

```python
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
categoria_tienda = pd.concat([tienda1, tienda2, tienda3, tienda4])
tabla_comparativa = categoria_tienda.groupby(['Categoría del Producto', 'Tienda'])['Producto'].count().unstack(fill_value=0)
tabla_comparativa.loc['TOTAL_TIENDA'] = tabla_comparativa.sum(axis=0)

print(tabla_comparativa)
```

# 3. Calificación promedio de la tienda


```python
# lista con DataFrames
tiendas = [tienda1, tienda2, tienda3, tienda4]

# Diccionario para almacenar los promedios
promedios = {}

# Iterar sobre la lista de DataFrames
for i, tienda in enumerate(tiendas):
    # Calculamos el promedio de la columna 'Calificación'
    promedio = tienda['Calificación'].mean()
    # Almacenamos el promedio en el diccionario, usando un nombre descriptivo
    promedios[f'Tienda {i+1}'] = promedio

# Imprimimos los promedios
for nombre_tienda, promedio in promedios.items():
    print(f'Promedio de Calificación en {nombre_tienda}: {promedio:,.2f}')
```

```python
# Convirtiendo el diccionario en series de panda:
series_calificacion_por_tienda = pd.Series(promedios)

# Así se puede graficar considerando las series:
series_calificacion_por_tienda.plot(kind='bar', figsize=(6, 4), color='blue')
```

# 4. Productos más y menos vendidos

```python
tabla_comparativa_masymenosvendidos = categoria_tienda.groupby(['Producto', 'Tienda'])['Producto'].count().sort_values(ascending=False)
#tabla_comparativa_masvendidos = categoria_tienda.groupby(['Producto', 'Tienda'])['Producto'].count().sort_values(ascending=False).unstack(fill_value=0)
print(tabla_comparativa_masymenosvendidos)

#print(tabla_comparativa_masvendidos)
#Para lograr que las tiendas aparezcan como columnas a la derecha (formato de matriz o tabla dinámica), la función  en Pandas es .unstack().
## 2. Desapilamos 'Tienda' para que pase de filas a columnas
# fill_value=0 es vital para que las tiendas que no vendieron un producto muestren 0
```

# 5. Envío promedio por tienda

```python
# Diccionario para almacenar los promedios
promedios_por_tienda = {}

# Iterar sobre la lista de DataFrames
for i, tienda in enumerate(tiendas):
    # Calculamos el promedio de la columna 'Calificación'
    promedio_envio = tienda['Costo de envío'].mean()
    # Almacenamos el promedio en el diccionario, usando un nombre descriptivo
    promedios_por_tienda[f'Tienda {i+1}'] = promedio_envio

# Imprimimos los promedios
for nombre_tienda, promedio_envio in promedios_por_tienda.items():
    print(f'Promedio del Costo de envío en {nombre_tienda}: {promedio_envio:,.0f}')
```

```python
# Convirtiendo el diccionario en series de panda:
series_promedios_por_tienda = pd.Series(promedios_por_tienda)

# Así se puede graficar considerando las series:
series_promedios_por_tienda.plot(kind='barh', figsize=(8, 6), color='purple')
```
